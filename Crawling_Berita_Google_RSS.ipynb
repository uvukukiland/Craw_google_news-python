```python
import feedparser
import pandas as pd
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Input keyword manual
keyword = input("Masukkan keyword berita yang ingin dicari: ").strip()
query = keyword.replace(" ", "+")

#  RSS Feed URL Google News
url = f"https://news.google.com/rss/search?q={query}&hl=id&gl=ID&ceid=ID:id"

# Parsing RSS feed
feed = feedparser.parse(url)

# Ekstrak berita
berita_list = []
for entry in feed.entries:
    judul = entry.title
    link = entry.link
    tanggal = entry.published
    sumber = entry.source.title if 'source' in entry else "Tidak diketahui"

    berita_list.append([judul, link, tanggal, sumber])

# Simpan ke DataFrame
df = pd.DataFrame(berita_list, columns=["Judul", "Link", "Tanggal", "Sumber"])

# Tampilkan DataFrame
print(f"âœ… Ditemukan {len(df)} berita untuk keyword: '{keyword}'")
display(df)

# Simpan ke file Excel
output_path = "/content/berita_google_news.xlsx"
df.to_excel(output_path, index=False)

# Tombol download
download_btn = widgets.Button(description="ðŸ“¥ Download Excel", button_style='success')

def on_download_clicked(b):
    with open(output_path, 'rb') as f:
        files.download(output_path)

download_btn.on_click(on_download_clicked)
display(download_btn)
```
